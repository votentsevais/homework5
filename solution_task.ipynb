```python
global documents
global directories
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

```python
#Унать владельца документа по его номеру
def p(n, k):
    for i in documents:
        if i['number'] == n:
            if k == 0:
                return'Владелец документа: ' + i['name']
            else:
                return i
    return 'Документ не найден в базе'
```

```python
#По номеру документа узнать на какой полке он хранится
def s(n, k):
    for i in directories:
        if n in directories[i]:
            if k == 0:
                return 'Документ хранится на полке: ' + i
            elif k == 1:
                return i
    return 'Документ не найден в базе'
```

```python
#увидеть полную информацию по всем документам
def l():
    for i in documents:
        k = s(i['number'], 1)
        print('№: ', i['number'], ', тип: ', i['type'], ', владелец: ', i['name'], ', полка хранения: ', k, sep='')
    return ''
```

```python
#Добавление новой полки
def ads(n):
    tt = ", ".join(directories.keys())
    if n in directories.keys():
        return 'Такая полка уже существует. Текущий перечень полок: ' + tt
    else:
        directories[n] = []
        return 'Полка добавлена. Текущий перечень полок: ' + tt + ', ' + str(n)
```

```python
# Удаление существующей полки из данных (только если она пустая)
def ds(n):
    if n in directories.keys():
        if directories[n] == []:
            del directories[n]
            return 'Полка удалена. Текущий перечень полок: ' + ", ".join(directories.keys())
        else:
            return 'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ' + ", ".join(directories.keys())
    else:
        return 'Такой полки не существует. Текущий перечень полок: ' + ", ".join(directories.keys())
```

```python
#Добавление нового документа в данные
def ad():
    n = input('Введите номер документа: ')
    t = input('Введите тип документа: ')
    p = input('Введите владельца документа: ')
    s = input('Введите полку для хранения: ')
    doc = {'type': t, 'number': n, 'name': p}
    if s in directories.keys():
        documents.append(doc)
        directories[s].append(n)
        print('Документ добавлен. Текущий список документов:')
    else:
        print('Такой полки не существует. Добавьте полку командой ads.')
    return l()
```

```python
# Удаление документа из данных
def d(n):
    i = p(n, 1)
    if i != 'Документ не найден в базе':
        documents.remove(i)
        directories[s(n, 1)].remove(n)
        print('Документ удален.')
    else:
        print(i)
    print('Текущий список документов:')
    l()
```

```python
cmd = input('Введите команду: ')
```

```python
while cmd != 'q':
    if cmd in ('p', 's', 'd'):
        n = input('Введите номер документа: ')
        if cmd == 'p':
            print(p(n, 0))
        elif cmd == 's':
            print(s(n, 0))
        elif cmd == 'd':
            d(n)
    elif cmd in ('ads', 'ds'):
        n = input('Введите номер полки: ')
        if cmd == 'ads':
            print(ads(n))
        else:
            print(ds(n))
    elif cmd == 'l':
        print(l())
    elif cmd == 'ad':
        print(ad())

    cmd = input('Введите команду: ')
else:
    print('Работа программы завершена')
```
